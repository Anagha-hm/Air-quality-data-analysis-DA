```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('C:/Users/Ankita/Desktop/5/DA_PROJ/city_hour.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
df.dropna()
```

```python
df.fillna(value=0, inplace=True)
```

```python
df.isnull().sum()
```

```python
df = df[df['Datetime'] >'2019-01-01 01:00:00']
```

```python
df
```

```python
plt.figure(figsize=(30,10))
plt.bar(df.groupby('City')['City'].count().index,df.groupby('City')['NO2'].mean(),color='red',alpha=0.6)
plt.xlabel("City",fontsize=8)
plt.ylabel("NO2 Values",fontsize=10)
plt.title("Air Quality NO2 in Indian Cities 2019-20")
plt.xticks(rotation=70)
```

```python
plt.figure(figsize=(30,10))
plt.bar(df.groupby('City')['City'].count().index,df.groupby('City')['PM2.5'].mean(),color='green',alpha=0.6)
plt.xlabel("City",fontsize=8)
plt.ylabel("PM2.5 Values",fontsize=10)
plt.title("PM2.5 in Indian Cities 2019-20")
plt.xticks(rotation=70)
```

```python
plt.figure(figsize=(30,10))
plt.bar(df.groupby('City')['City'].count().index,df.groupby('City')['AQI'].mean(),color='brown',alpha=0.9)
plt.xlabel("City",fontsize=8)
plt.ylabel("Air Quality Index",fontsize=10)
plt.title("AQI in Indian Cities 2019-20")
plt.xticks(rotation=70)
```
